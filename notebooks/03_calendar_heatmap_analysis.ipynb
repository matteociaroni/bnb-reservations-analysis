# Calendar heatmap analysis
Visualize daily metrics across the year.

## Load data
Load data from the CSV files computed using the prepare_data.py script.

```python
import sys
import pandas as pd
import numpy as np
from pathlib import Path
sys.path.append(str(Path("..").resolve()))
from src.plots import build_day_of_year_heatmap, plot_calendar_heatmap, plot_daily_bookings_by_platform
import seaborn as sns

df_daily = pd.read_csv("../output/daily_bookings.csv", parse_dates=["date"])
monthly_summary = pd.read_csv("../output/monthly_summary.csv")
years = sorted(monthly_summary["year"].unique())
months = np.arange(1, 13)
month_labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
```

## Graphs

```python
plot_daily_bookings_by_platform(df_daily, year=2024)
```

### Occupied days

```python
heatmap_booked_years = build_day_of_year_heatmap(
    df_daily=df_daily,
    value_col="year",
    agg_func=pd.Series.nunique
)

plot_calendar_heatmap(
    heatmap_booked_years,
    title="Booked calendar days",
    colorbar_label="Number of years booked",
    cmap="YlGnBu"
)
```
