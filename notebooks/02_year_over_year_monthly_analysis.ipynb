# Monthly year-over-year analysis
In this notebook, we compute monthly performance metrics and compare them year by year to identify trends and seasonality.

## Load data
Load data from the CSV files computed using the prepare_data.py script.

```python
import sys
import pandas as pd
import numpy as np
from pathlib import Path
sys.path.append(str(Path("..").resolve()))
from src.plots import plot_monthly_metric_by_year

df_daily = pd.read_csv("../output/daily_bookings.csv", parse_dates=["date"])
monthly_summary = pd.read_csv("../output/monthly_summary.csv")
years = sorted(monthly_summary["year"].unique())
months = np.arange(1, 13)
month_labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
```

## Function for creating monthly graph

## Graphs
### Nights per month

```python
plot_monthly_metric_by_year(
    df=monthly_summary,
    value_column="occupied_nights",
    years=years,
    months=months,
    month_labels=month_labels,
    title="Occupied nights per month",
    y_label="Nights"
)
```

### Occupancy rate

```python
plot_monthly_metric_by_year(
    df=monthly_summary,
    value_column="occupancy_pct",
    years=years,
    months=months,
    month_labels=month_labels,
    title="Monthly occupancy rate",
    y_label="Occupancy (%)",
    ylim=(0, 100)
)
```

### Average number of persons per night

```python
plot_monthly_metric_by_year(
    df=monthly_summary,
    value_column="avg_persons",
    years=years,
    months=months,
    month_labels=month_labels,
    title="Average persons per night",
    y_label="Persons"
)
```
